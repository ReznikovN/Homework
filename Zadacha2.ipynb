```python
import math
import matplotlib.pyplot as plt

# начальные параметры цикла
P_0 = 5  # давление на входе в турбину, МПа
T_0 = 450 + 273.15  # температура на входе в турбину, К
P_2_list = [5, 10, 15, 20, 50]  # давление на выходе из турбины, кПа


def eta_cycle(P_2):
    P_2 = P_2 * 1000  # переводим давление в Па
    P_1 = P_0 * 10**6  # переводим давление в Па
    T_1 = T_0
    T_3 = 60 + 273.15  # температура на выходе из конденсатора, К
    c_p = 4400  # удельная теплоемкость пара, Дж/(кг*К)
    c_v = 3100  # удельная теплоемкость при постоянном объеме, Дж/(кг*К)
    k = c_p / c_v
    eta_th = 1 - (P_2/P_1)**((k-1)/k)

    return eta_th


# значения для оси x
x_values = [P_2 for P_2 in P_2_list]

# значения для оси y
y_values = [eta_cycle(P_2) for P_2 in P_2_list]

# построение графика
plt.plot(x_values, y_values, marker='o')
plt.xlabel('P_2, кПа')
plt.ylabel('η_th')
plt.title('Термический КПД цикла без промежуточного перегрева пара')
plt.show()
```
